```
import matplotlib.pyplot as plt
#matplotlib - Package(collection of libraries)
#pyplot- Sub-package(library)
```

```
#Line plot
x=[1,2,-6,9]
y=[5,-7,0,-5]

plt.plot(x,y,c='w')
ax=plt.axes()
ax.set_facecolor('k')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('LINE PLOT')
plt.show()
```

```
#SCATTERPLOT
import numpy as np
x=np.random.randint(1,100,50)
y=np.random.randint(1,100,50)

plt.scatter(x,y,c='w')
ax=plt.axes()
ax.set_facecolor('k')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('SCATTER PLOT')
plt.show()
```

```
#LINExSCATTER plot
a = np.random.randint(1,10,9)
b =np.arange(21,30,1)
print(a,b)
plt.plot(a,b) # Predicted Line
plt.scatter(a,b,c = 'r') # Original Data points
plt.show()
```

```
# Bar Graph
name = [ 'sam','ram','nam','yam']
weight = [42,65,36,58]
height = [123,145,167,178]
c_list = ['red' if x>45 else 'blue' for x in weight] # List Comprehension
print(c_list)
plt.bar(name,weight,width=0.4,color = c_list)
#plt.bar(name,height,width=0.8,color = 'b')
plt.xlabel('Person')
plt.ylabel('Weight')
plt.title('Bar Graph : Person vs Weight')
plt.show()
```

```
elemen=['negateium','moderateium','heavium']
weigh=[-200,150,450]
plt.bar(elemen,weigh,color=['r','blue','k'])
plt.xlabel('ELEMENTS')
plt.ylabel('WEIGHT')
plt.show()
```
